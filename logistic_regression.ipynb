```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sig(z):
    return (1/(1+np.exp(-z)))
```

```python
n =100

X = np.linspace(-100,100,n)
Y=[]
for i in X:
    if i<0:
        Y.append(0)
    elif -20<=i<=20:
        lab=np.random.randint(2)
        Y.append(lab)
    elif 20<i<=100:
        Y.append(1)
    
Y=np.array(Y)
Y.size
```

```python
m ,c = 0.5, -2.0
lr = 0.1
iteration = 10000
```

```python
plt.scatter(X,Y,marker='x',c="black")
plt.plot(X,sig(m*X+c))
```

```python
M_hist,C_hist = np.zeros(iteration),np.zeros(iteration)
for i in range(iteration):
    M_hist[i] = m
    C_hist[i] = c
    z = m*X+c
    y_hat = sig(z)

    error = y_hat-Y
    dm = np.mean(error*X)
    dc = np.mean(error)

    m -= lr*dm
    c -= lr*dc

print(f"slope: {m:.2f}\nintercept: {c:.2f}")    
```

```python
plt.scatter(X,Y,marker='x',c="black")
plt.plot(X,sig(m*X+c))
```

```python
hours = np.array([3.5, 5.5, 7.0])
probs = sig(m * hours + c)
for h, p in zip(hours, probs):
    print(f"{h} hours → {p:.2f} probability → {'PASS' if p > 0.5 else 'FAIL'}")
```
