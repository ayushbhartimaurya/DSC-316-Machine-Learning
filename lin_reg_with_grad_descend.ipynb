```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
n = 100

x = np.random.random(n)
y = np.random.random(n)
```

```python
plt.scatter(x,y,marker='x',c='black')
```

```python
m = 1
c = 1
lr = 0.1
iteration = 10000
```

```python
for _ in range(iteration):
    y_hat = m*x+c
    error = y_hat-y

    dm = np.mean(error*x)
    dc = np.mean(error)

    m -= lr*dm
    c -= lr*dc

print(f"m : {m}\nc : {c}")
```

```python
plt.scatter(x,y,marker='x',c='black')
plt.plot(x,m*x+c, c='red')
```
