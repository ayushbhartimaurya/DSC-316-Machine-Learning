# Linear Regression with Gradient Descent

This notebook implements linear regression and fits the parameters using gradient descent.

## Key definitions
- Linear regression models a continuous target as a weighted sum of input features.
- Gradient descent minimizes the error between predictions and targets.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Generating sample data

Creating a random dataset of paired $x$ and $y$ values for the regression line.

```python
n = 100

x = np.random.random(n)
y = np.random.random(n)
```

## Visualizing the scatter plot

Plotting the raw points before fitting the line.

```python
plt.scatter(x,y,marker='x',c='black')
plt.show()
```

## Setting the initial parameters

Initializing the slope, intercept, learning rate, and iteration count.

```python
m = 1
c = 1
lr = 0.1
iteration = 10000
```

## Running gradient updates

Applying $m \leftarrow m - \alpha\ * dm$ and $c \leftarrow c - \alpha\ * dc$ repeatedly.

```python
for _ in range(iteration):
    y_hat = m*x+c
    error = y_hat-y

    dm = np.mean(error*x)
    dc = np.mean(error)

    m -= lr*dm
    c -= lr*dc

print(f"m : {m}\nc : {c}")
```

## Plotting the fitted line

Overlaying the learned regression with aforementioned slope and intercept line on top of the scatter plot.

```python
plt.scatter(x,y,marker='x',c='black')
plt.plot(x,m*x+c, c='red')
plt.show()
```
