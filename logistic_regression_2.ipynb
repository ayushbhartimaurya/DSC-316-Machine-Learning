```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
n = len(y)
m = [1,1,1]
c= 1
lr = 0.1

iteration = 100000
```

### Training Loop

```python
for _ in range(iteration):
    z = x@m +c
    y_hat = sigmoid(z)
    error = y_hat -y
    dm = x.T@error/n
    dc = np.mean(error)

    m -= lr*dm
    c -= lr*dc

print(f"m = {m},\nc = {c:.2f}")
```

### Testing

```python
studentA = np.array([[9,64,60],[2,10,42],[9,10,90],[1,10,90]])
probs= sigmoid(studentA@m+c)
for s, p in zip(studentA, probs):
    print(f"student with stats {s} → {p:.2f} probability → {'PASS' if p > 0.5 else 'FAIL'}")

preds = (sigmoid(x @ m + c) > 0.5).astype(int)
accuracy = np.mean(preds == y) * 100
print(f"Accuracy: {accuracy:.1f}%")
```
