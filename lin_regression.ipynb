# Linear Regression

This notebook demonstrates linear regression on a small random example dataset.

## Key definitions
- Linear regression predicts a continuous value using a linear combination of inputs.
- The fitted line is chosen to minimize prediction error.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 100

X = np.random.random(n)
Y = np.random.random(n)
```

```python
m = (n*np.sum(X*Y)-(np.sum(X)*np.sum(Y)))/(n*np.sum(X**2)-(np.sum(X))**2)

c = (np.sum(Y)-m*np.sum(X))/n
```

```python
print(f"slope is {m:2f}")
print(f"intercept is {c:2f}")
```

```python

plt.scatter(X,Y,marker="+", c="black")
plt.plot(X,m*X+c,c='skyblue')
plt.show()
```
