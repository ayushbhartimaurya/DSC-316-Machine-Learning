```python
import numpy as np
```

```python
X = np.array([
    [1,20],[2,40],[3,55],[4,60],
    [6,70],[7,80],[8,85],[9,90],
    [2,35],[5,65],[7,72],[3,48],
    [8,88],[1,25],[6,68],[4,58]
], dtype=float)
y = np.array([0,0,0,1,1,1,1,1,0,1,1,0,1,0,1,1], dtype=float)
```

```python
X = (X-np.mean(X))/np.std(X)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
np.random.seed(42)
w = np.zeros(X.shape[1])
b = 0
lr = 0.1
n = len(y)
iterations = 100000
noise = 1e-8

losses = []
```

```python
for epoch in range(iterations):
    z = X@w+b
    y_hat = sigmoid(z)

    loss = -np.mean(y * np.log(y_hat + noise) + (1-y) * np.log(1-y_hat + noise))
    losses.append(loss)

    error = y_hat - y

    dw = X.T@error/n
    db = np.mean(error)

    w = w - lr*dw
    b = b - lr*db

    if epoch % 10000 == 0:
        preds = (y_hat > 0.5).astype(int)
        acc = np.mean(preds == y) * 100
        print(f"epoch {epoch:4d} | loss: {loss:.4f} | accuracy: {acc:.1f}%")

print(f"\nFinal weights: {w}")
print(f"Which feature matters more? {'hours' if abs(w[0]) > abs(w[1]) else 'prev_score'}")
```
