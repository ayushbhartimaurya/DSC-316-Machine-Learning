# Gradient Descent

This notebook demonstrates gradient descent as an optimization method for machine learning models.

## Key definitions
- Gradient descent updates parameters in the direction that reduces the loss.
- It does so by moving towards the negative gradient 
- The learning rate controls the size of each update step.

```python
import numpy as np
import pandas as pd
```

## Building a sample dataset

defining a custom arbitrary dataset

```python
X = np.array([
    [1,20],[2,40],[3,55],[4,60],
    [6,70],[7,80],[8,85],[9,90],
    [2,35],[5,65],[7,72],[3,48],
    [8,88],[1,25],[6,68],[4,58]
], dtype=float)
y = np.array([0,0,0,1,1,1,1,1,0,1,1,0,1,0,1,1], dtype=float)

pd.DataFrame({'Hours_studied': X[:, 0], 'Prev_score': X[:, 1], 'Passed': y.astype(bool)})
```

## normalizing the inputs

Normalizing the inputs such that their mean is 0 and standar deviation is 1

$$\text{Z} = \frac{X - \mu}{\sigma^2}$$

- $\mu$ = mean of X

- $\sigma$ = standard deviation of X

```python
X = (X-np.mean(X))/np.std(X)
```

## Defining the sigmoid function

Sigmoid function $\sigma(z)$ converts linear scores into probabilities.

$$\sigma(z)=\frac{1}{1+e^{-z}}$$

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

## Initializing training variables

Setting the weights, bias, learning rate, and loss history used in gradient descent.

```python
np.random.seed(42)
w = np.zeros(X.shape[1])
b = 0
lr = 0.1
n = len(y)
iterations = 100000  # Iterations for gradient descent
noise = 1e-8

losses = []
```

## Running gradient descent 

Each epoch computes $z=X*w+b$, applies the sigmoid, evaluates binary cross-entropy loss, and updates $w$ and $b$ with the gradients.

```python
for epoch in range(iterations):
    z = X@w+b
    y_hat = sigmoid(z)

    loss = -np.mean(y * np.log(y_hat + noise) + (1-y) * np.log(1-y_hat + noise))
    losses.append(loss)

    error = y_hat - y

    dw = X.T@error/n
    db = np.mean(error)

    w = w - lr*dw
    b = b - lr*db

    if epoch % 10000 == 0:
        preds = (y_hat > 0.5).astype(int)
        acc = np.mean(preds == y) * 100
        print(f"epoch {epoch:4d} | loss: {loss:.4f} | accuracy: {acc:.1f}%")

print(f"\nFinal weights: {w}")
print(f"Which feature matters more? {'hours' if abs(w[0]) > abs(w[1]) else 'prev_score'}")
```
